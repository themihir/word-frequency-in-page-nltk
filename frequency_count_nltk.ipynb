```python
import nltk
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
import urllib.request
import plotly.io as pio
```

```python
enter_page = 'https://en.wikipedia.org/wiki/Machine_learning'
page =  urllib.request.urlopen(enter_page)
```

```python
html_plain = page.read()
```

```python
soup = BeautifulSoup(html_plain,'html.parser')
soup_text = soup.get_text(strip = True)
```

```python
ready_text = soup_text.lower()
print(ready_text)
tokens = []
for t in ready_text.split():
    tokens.append(t)
print(tokens)
```

```python
stop_words = stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stop_words:
        clean_tokens.remove(token)
print(clean_tokens)
```

```python
freq = nltk.FreqDist(clean_tokens)
for word, count in freq.items():
    print('Word: ', word, ', Qunatity:', count)
```

```python
high_freq = {}
for key, val in freq.items():
    if(val>10):
        high_freq[key]=val
```

```python
fig = dict({
    "data": [{"type": "bar",
              "x": list(high_freq.keys()),
              "y": list(high_freq.values())}],
    "layout": {"title": {"text": "Most frequently used words in the page"}, "xaxis": {"categoryorder":"total descending"}}
})
```

```python
pio.show(fig)
```
